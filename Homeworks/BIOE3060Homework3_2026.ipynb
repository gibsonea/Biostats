Most medical screening tests are not perfectly accurate. The accuracy is typically measured using two conditional probabilities:

-   The <font color="tomato">sensitivity</font> of the test is the probability that a person with the disease correctly receives a positive test. This is a <font color="tomato">Correct Positive</font>.

$$\color{tomato}{\mbox{Sensitivity} = P\left( + \mbox{ Test } \big| \mbox{ Has Disease} \right)}$$

-   The <font color="dodgerblue">specificity</font> of the test is the probability that a person without the disease correctly receives a negative test. This is a <font color="dodgerblue">Correct Negative</font>.

$$\color{dodgerblue}{\mbox{Specificity} = P( - \mbox{ Test } \big| \mbox{ No Disease})}$$

This problem will focus on the Mini Mental State Examination (MMSE), which is the most commonly used test for complaints of problems with memory or other mental abilities<sup>1</sup>.

-   <font color="tomato">The sensitivity of the MMSE is 80%, $P\left( + \mbox{ Test } \big| \mbox{ Has Disease} \right) = 0.8$</font>.
-   <font color="dodgerblue">The specificity of the MMSE is 90%, $P\left( - \mbox{ Test } \big| \mbox{ No Disease} \right) = 0.9$</font>.

## <a name="HW3Q3a">Question 3a</a>

Use the rnorm command to generate a dataset of values for Xd from a diagnostic test. The healthy population, denoted by H, will have random values drawn from a distribution with a mean of 5 and standard deviation of 0.6. The disease population, denoted by the vector D, will have a mean of 5.5 and standard deviation of 0.5.

Fill in the question marks of the code cell below and run the code.

```r
set.seed(2529)  #this value seeds the random number generator

H <- rnorm(20, mean = ?, sd = ?)
D <- rnorm(20, mean = ?, sd = ?)


xmin <- min(H, D)
xmax <- max(H, D)


# create a histogram
hist(H,  # vector of values to plot
     breaks = 10,  # number of bin ranges to use
     xlim = c(xmin, xmax),
     xlab = "Xd",   # x-axis label
    ylab = "Frequency",  # y-axis label
main = "Distribution of Healthy and Disease groups",  # main label
    cex.lab=1.5, cex.axis=1.5, cex.main=1.5,  # increase font size
     col = "steelblue")  # fill color of bars

hist(D, col = "red", breaks = 10,
     add = TRUE)

legend("topright", c("Healthy", "Disease"), fill=c("steelblue", "red"))

```

Execute the code below to plot the ROC curve.

```r
#First you need to put the data into a single vector of Xd values

Xd = c(H,D) #concatenate two vectors

#Next you need to create a vector specifying the category healthy or disease
group <- c(rep('Healthy', length(H)), rep('Disease', length(D)))

#This command calculates and plots the curve
plot.roc(group, Xd,
                    main="ROC", percent=FALSE, col="#1c61b6")

```

Execute the code below. The command roc calculates the parameters for the ROC and stores these parameters in a Dataframe.


```r
roc1 = roc(group, Xd)
```
